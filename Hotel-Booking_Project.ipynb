# Data analysis and Visualization

```python
# Calculate cancellation percentage
cancelled_perc = df["is_canceled"].value_counts(normalize=True)
print(cancelled_perc)

# Define labels properly
labels = ["Not Canceled", "Canceled"]
counts = df["is_canceled"].value_counts()

# Plot the bar chart
plt.figure(figsize=(5,4))
plt.title("Reservation Status Count")
plt.bar(labels, counts, edgecolor="k", width=0.7, color=["green", "red"])  # Add colors for clarity
plt.ylabel("Number of Reservations")
plt.show()
```

```python
plt.figure(figsize=(8, 4))
ax1 = sns.countplot(x='hotel', hue='is_canceled', data=df, palette='Blues')

# Adjust legend position
ax1.legend(title="Cancellation Status", labels=["Not Canceled", "Canceled"], bbox_to_anchor=(1, 1))

# Add title and labels
plt.title('Reservation Status in Different Hotels', size=20)
plt.xlabel('Hotel')
plt.ylabel('Number of Reservations')

# Show plot
plt.show()
```

```python
resort_hotel = df[df['hotel'] == 'Resort Hotel']
# Calculate cancellation rate for Resort Hotel
resort_hotel_cancellation_rate = resort_hotel['is_canceled'].value_counts(normalize=True)
print(resort_hotel_cancellation_rate)
```

```python
# Filter the City Hotel data
city_hotel = df[df['hotel'] == 'City Hotel']

# Calculate cancellation rate for City Hotel
city_hotel_cancellation_rate = city_hotel['is_canceled'].value_counts(normalize=True)
print(city_hotel_cancellation_rate)
```

```python
# Group by 'reservation_status_date' and calculate mean ADR for Resort Hotel
resort_hotel_grouped = resort_hotel.groupby('reservation_status_date')[['adr']].mean()

# Group by 'reservation_status_date' and calculate mean ADR for City Hotel
city_hotel_grouped = city_hotel.groupby('reservation_status_date')[['adr']].mean()

# Plot the average daily rate (ADR) for both Resort and City Hotel
plt.figure(figsize=(20, 8))

plt.title('Average Daily Rate in City and Resort Hotel', fontsize=30)

# Plot for Resort Hotel
plt.plot(resort_hotel_grouped.index, resort_hotel_grouped['adr'], label='Resort Hotel', color='b')

# Plot for City Hotel
plt.plot(city_hotel_grouped.index, city_hotel_grouped['adr'], label='City Hotel', color='g')

plt.xlabel('Reservation Status Date', fontsize=15)
plt.ylabel('Average Daily Rate (ADR)', fontsize=15)
plt.legend()
plt.xticks(rotation=45)  # Rotate x-axis labels for readability

plt.show()
```

```python
# Extract month from 'reservation_status_date'
df['month'] = df['reservation_status_date'].dt.month

# Create the countplot to visualize reservation status per month
plt.figure(figsize=(16, 8))

# Corrected countplot function syntax
ax1 = sns.countplot(x='month', hue='is_canceled', data=df, palette='bright')

# Fixing the legend handling and getting the labels
legend_labels, _ = ax1.get_legend_handles_labels()

# Adjusting the legend to be outside the plot
ax1.legend(bbox_to_anchor=(1, 1))

# Set the plot title and axis labels
plt.title('Reservation Status per Month', fontsize=20)
plt.xlabel('Month', fontsize=15)
plt.ylabel('Number of Reservations', fontsize=15)

# Adding a custom legend
plt.legend(['Not Canceled', 'Canceled'])

# Display the plot
plt.show()
```

```python
plt.figure(figsize=(15, 8))

# Set the title
plt.title('ADR per month', fontsize=30)

# Create a barplot for ADR per month for canceled reservations only
sns.barplot(x='month', y='adr', data=df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index())

# Show the legend
plt.legend(fontsize=20)

# Display the plot
plt.show()
```

```python
# Filter the canceled data
cancelled_data = df[df['is_canceled'] == 1]

# Get the top 10 countries with canceled reservations
top_10_country = cancelled_data['country'].value_counts()[:10]

# Create a pie chart for the top 10 countries
plt.figure(figsize=(8, 8))

# Set the title for the plot
plt.title('Top 10 countries with reservation canceled')

# Create the pie chart
plt.pie(top_10_country, autopct='%.2f%%', labels=top_10_country.index)

# Display the plot
plt.show()
```

```python
# Filter the rows where the reservations are canceled (is_canceled == 1)
cancelled_df = df[df['is_canceled'] == 1]

# Filter the rows where the reservations are not canceled (is_canceled == 0)
not_cancelled_df = df[df['is_canceled'] == 0]

# Group by 'reservation_status_date' and calculate the mean of 'adr' for canceled reservations
cancelled_df_adr = cancelled_df.groupby('reservation_status_date')[['adr']].mean()

# Reset index for better plotting
cancelled_df_adr.reset_index(inplace=True)

# Sort values by 'reservation_status_date'
cancelled_df_adr.sort_values('reservation_status_date', inplace=True)

# Group by 'reservation_status_date' and calculate the mean of 'adr' for not canceled reservations
not_cancelled_df_adr = not_cancelled_df.groupby('reservation_status_date')[['adr']].mean()

# Reset index for better plotting
not_cancelled_df_adr.reset_index(inplace=True)

# Sort values by 'reservation_status_date'
not_cancelled_df_adr.sort_values('reservation_status_date', inplace=True)

# Plot the data
plt.figure(figsize=(20, 6))
plt.title('Average Daily Rate')
plt.plot(not_cancelled_df_adr['reservation_status_date'], not_cancelled_df_adr['adr'], label='Not Cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label='Cancelled')
plt.legend()
plt.show()
```
